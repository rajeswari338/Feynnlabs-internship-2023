```python
from sklearn.decomposition import PCA
# List of columns used for PCA
columns_for_pca = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap',
       'tasty', 'expensive', 'healthy', 'disgusting']

data_for_pca = data[columns_for_pca]
pca = PCA()

pca.fit(data_for_pca)

variance_explained = pca.explained_variance_ratio_

cumulative_variance = pca.explained_variance_ratio_.cumsum()

print("\nProportion of Variance Explained:")
print(variance_explained)

print("\nCumulative Proportion of Variance Explained:")
print(cumulative_variance)
```

```python
std_deviation = pca.explained_variance_ ** 0.5
```

```python
print("Standard Deviation of Components:")
print(std_deviation)
```

```python


std_deviation = pca.explained_variance_ ** 0.5
rotation_matrix = pca.components_.T

# Print standard deviations
print("Standard deviations (1, .., p={0}):".format(len(columns_for_pca)))
print(np.array2string(std_deviation, separator=',', formatter={'float_kind':lambda x: "%.1f" % x}))

# Print rotation matrix
print("Rotation (n x k) = ({0} x {1}):".format(len(columns_for_pca), len(columns_for_pca)))
print(np.array2string(rotation_matrix, separator=',', formatter={'float_kind':lambda x: "%.3f" % x}))
```

```python

data_pca = pca.transform(data_for_pca)

# Plot the PCA projection
plt.scatter(data_pca[:, 0], data_pca[:, 1], c="grey")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Projection")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

data = np.random.rand(100, 10)

# Perform PCA
pca = PCA()
pca.fit(data)

pc_list = list(range(1, len(pca.explained_variance_ratio_) + 1))

# Create a cumulative explained variance plot (Scree plot)
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
plt.figure(figsize=(10, 5))
plt.plot(pc_list, cumulative_variance, marker='o', linestyle='-', color='b')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Scree Plot')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=1234)
    kmeans.fit(data_pca)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()


```

```python
optimal_k = 4 

# Perform KMeans clustering
kmeans = KMeans(n_clusters=optimal_k, random_state=1234)
cluster_labels = kmeans.fit_predict(data_pca)

# Plot the PCA projection with cluster colors
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=cluster_labels, cmap='viridis')
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Projection with Clusters")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
num_columns = data.shape[1]
# Calculate the number of rows and columns for subplots
num_rows = int(np.ceil(np.sqrt(num_columns)))
num_cols = int(np.ceil(num_columns / num_rows))

plt.rcParams['figure.figsize'] = (12, 14)

fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 14))
fig.subplots_adjust(hspace=0.5) 


for col, ax in enumerate(axes.flatten()):
    if col < num_columns:
        ax.hist(data[:, col], bins=10)  
        ax.set_title(f"Histogram for Column {col}")
        ax.set_xlabel("Value")
        ax.set_ylabel("Frequency")
    else:
        ax.axis('off') 


plt.tight_layout()
plt.show()
```
