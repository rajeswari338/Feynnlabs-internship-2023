```python
from sklearn.cluster import KMeans
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++",random_state=28)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))    
plt.grid()
plt.plot(range(1,11),wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("WCSS")
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Select categorical variables for segmentation
categorical_vars = ['Profession', 'Marrital Status', 'Education', 'Personal loan', 'House Loan', 'Wife Working', 'Age']

# Select the target variable for segmentation
target_var = 'Make'  # You can adjust this based on your segmentation criteria

# Initialize an empty list to store variable importance scores
var_importance = []

# Iterate through each categorical variable and calculate chi-squared statistic
for var in categorical_vars:
    contingency_table = pd.crosstab(df[var], df[target_var])
    chi2, _, _, _ = chi2_contingency(contingency_table)
    var_importance.append((var, chi2))

# Sort variables by chi-squared statistic (importance) in descending order
var_importance.sort(key=lambda x: x[1], reverse=True)

# Select the top 4 variables
top_4_variables = [var for var, _ in var_importance[:4]]

# Display the top 4 variables
print("Top 4 Variables for Market Segmentation:")
for var in top_4_variables:
    print(var)
```

```python


# Basic segmentation based on criteria mentioned
# You can add more detailed segmentation as needed

# Age segmentation
df['Age Group'] = pd.cut(df['Age'], bins=[20, 30, 40, 50], labels=['Young Adults', 'Middle-Aged', 'Senior'])

# Financial segmentation
df['Income Level'] = pd.cut(df['Total Salary'], bins=[0, 1200000, 1800000, float('inf')], labels=['Low Income', 'Middle Income', 'High Income'])

# Number of Dependents segmentation
df['Dependents Group'] = pd.cut(df['No of Dependents'], bins=[0, 1, 2, float('inf')], labels=['No Dependents', '1-2 Dependents', '3+ Dependents'])



# You can add more segments and perform additional analysis as required.
```
